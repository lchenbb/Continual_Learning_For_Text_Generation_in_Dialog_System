```python
import numpy as np
import torch 
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib
import os
from collections import Counter, OrderedDict
import pandas as pd
from IPython.display import display, HTML
import numbers

import seaborn as sns
sns.set()

```

```python

base_folders = './experiments/da/'

#exp_folders = [ 'finetune', 'examplar_random_128', 'examplar_loss_128', 'examplar_random_distillation_128', 'examplar_loss_distillation_128', 'upper_bound']
raw_records = {}
#seed_folders = ['1111/', '2222/', '3333/', '4444/', '5555/']
seed_folders = ['1111/']

for seed_folder in seed_folders:
    
    raw_records[seed_folder] = {}
    for exp_folder in sorted(os.listdir(base_folders + seed_folder)):
    
        if exp_folder != '.DS_Store' and exp_folder != 'curve':
            print (exp_folder)
            path = base_folders + seed_folder + exp_folder + '/model/'
            raw_records[seed_folder][exp_folder] = []
            for file in sorted(os.listdir(path)):

                #if file != '.DS_Store' and '023541' in file :
                if file != '.DS_Store':
                    ckp = torch.load(path+file,map_location='cpu')['perf']
                    raw_records[seed_folder][exp_folder].append(ckp)

    #print (sv_records[exp_folder][-1])
#print(raw_records)
```

```python
def tabulate_results(records, metric='cul_bleu'):
    '''
    tabulate results for each task/domain
    '''
    formated_results = []
    all_results = {}
    for seed, results in records.items():
        for method, perf in results.items():

            result = []
            for run in perf:
                if metric == 'cul_bleu':
                    result.append(run[metric][3])
                elif metric == 'bleu':
                    result.append(run[metric][0][3])
                elif 'cul' in metric:
                    result.append(run[metric])
                else:
                    result.append(run[metric][0])
            if method in all_results.keys():
                all_results[method] += [result]
            else:
                all_results[method] = [result]
    
    #print(all_results)
    filtered_results = {}
    for method, result in all_results.items():
        if len(result) == len(seed_folders):
            filtered_results[method] = result
    
    for method, result in filtered_results.items():
        result = np.array(result)
        mean_result = list(result.mean(axis=0))
        std_result = list(result.std(axis=0))
    
        formated_results.append( [method] + mean_result )

    headers = ['methods with: '+metric, 'Attraction','Train','Hotel','Booking', 'Restaurant', 'Taxi', 'ds']
    
    print(formated_results)
    return formated_results, headers
```

```python
import numpy as np
def tabulate_results_all(records):
    '''
    tabulate results for all, base, new
    '''
    
    formated_results = []
    all_results = {}
    for seed, results in records.items():
        for method, perf in results.items():

            result = []
            sequence = [5,3,2,0,1,4]
            sequence = [0,5,2,1,3,4]
            sequence = [0,2,7,3,6,4,1]
            ind = 0
            for run in perf:
                # all
                result.append(run['cul_se'])
                result.append(run['cul_bleu'][3])
                # base (the first domain: 0)
                result.append(run['se'][0])
                result.append(run['bleu'][0][3])
                # new
                result.append(run['se'][sequence[ind]])
                result.append(run['bleu'][sequence[ind]][3])               
                ind += 1

            result = np.asarray(result).reshape(7,6).T.mean(1)
            
            if method in all_results.keys():
                all_results[method] += [result]
            else:
                all_results[method] = [result]
    
    #print(all_results)
    filtered_results = {}
    for method, result in all_results.items():
        if len(result) == len(seed_folders):
            filtered_results[method] = result
    
    for method, result in filtered_results.items():
        result = np.array(result)
        mean_result = list(result.mean(axis=0))
        std_result = list(result.std(axis=0))
    
        formated_results.append( [method] + mean_result )

    headers = ['Methods', 'ALL SER', 'ALL BLEU', 'Base SER','Base BLEU' , 'New SER','New BLEU']
    
    return formated_results, headers
```

```python
results, headers = tabulate_results_all(raw_records)
dataframe = pd.DataFrame(results, columns=headers)
#dataframe.to_csv ('./experiments/domain/seed/bleu.csv', index = False, header=True)
display(HTML(dataframe.to_html()))
```

```python
def plot_results(records, metric = 'cul_loss'):
    import matplotlib.pyplot as plt
    import matplotlib
    import seaborn as sns
    sns.set()
    sns.set_color_codes(palette='bright')
    
    plt.figure(figsize=(9,7.5))
    matplotlib.style.use('seaborn')
    
    plt.rc('axes', titlesize=28)  # fontsize of the axes title
    plt.rc('axes', labelsize=25)  # fontsize of the x and y labels
    plt.rc('xtick', labelsize=22)  # fontsize of the tick labels
    plt.rc('ytick', labelsize=18)  # fontsize of the tick labels
    plt.rc('legend', fontsize=16)  # legend fontsize
    
    for seed, results in records.items():
        for method, perf in results.items():
            result = []
            for run in perf:
                if metric == 'cul_bleu':
                    result.append(run[metric][3])
                elif metric == 'bleu':
                    result.append(run[metric][0][3])
                elif 'cul' in metric:
                    result.append(run[metric])
                else:
                    result.append(run[metric][0])
            print (method)
            print (result)

            x = ['Attraction','Hotel','Restaurant','Train','Taxi', 'Booking']
            plt.plot(x, result, label=method)

#     plt.gca().set_ylim([3.5,68])
#     plt.gca().set_yticklabels(['{:.0f}%'.format(x) for x in plt.gca().get_yticks()])
#    plt.gca().set_xticks(['Attraction','Hotel','Rest','Train','Taxi', 'Booking'])

    plt.xticks(rotation=45)
    plt.ylabel(metric)
    #plt.xlabel('Domains')
    plt.title('MultiWoz NLG',fontweight='bold')
    plt.legend(loc=0,frameon=False)
    plt.tight_layout()
    plt.show()
    
```
