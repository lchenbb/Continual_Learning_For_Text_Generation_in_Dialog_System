```python
from nltk.translate.bleu_score import corpus_bleu, SmoothingFunction

def get_bleu(refs, gens):

    refs = [[s[0].split()] for s in refs]
    gens = [s[0].split() for s in gens]
    
#     print(refs)
#     print(gens)
    smooth = SmoothingFunction()
    BLEU = []
    weights = [(1, 0, 0, 0), (0.5, 0.5, 0, 0), (0.333, 0.333, 0.333, 0), (0.25, 0.25, 0.25, 0.25)]
    for i in range(4):
        bleu = corpus_bleu(refs, gens, weights=weights[i],
                           smoothing_function=smooth.method1)
        print(bleu)
        BLEU.append(bleu)
        
    print('BLEU 4:', BLEU)

    return BLEU

ref = []
gen = []
with open('sclstm.res') as f:
    for line in f:
        if 'Target' in line:
            target = line.strip().split(':')[1][1:]
            ref.append([target])
            
        if 'Gen' in line:
            g = line.strip().split(':')[1][1:]
            gen.append([g])

get_bleu(ref, gen)
```
